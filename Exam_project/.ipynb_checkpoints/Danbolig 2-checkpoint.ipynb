```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time
import threading

# Function to save HTML content to a text file
def save_html_to_file(html_content, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(html_content)

# Create a new instance of the Chrome driver
driver = webdriver.Chrome()

# Open the URL
url = 'https://danbolig.dk/soeg/?o=newest-desc'
driver.get(url)

# Function to periodically save HTML content
def save_html_periodically():
    while True:
        # Get the HTML content of the current page
        page_source = driver.page_source
        timestamp = time.strftime('%Y%m%d%H%M%S')
        filename = f'page_{timestamp}.html'
        
        # Save the HTML content to a text file
        save_html_to_file(page_source, filename)
        
        print(f'Saved HTML content to {filename}')
        time.sleep(300)  # Wait for 5 minutes

# Start the thread to save HTML content periodically
html_thread = threading.Thread(target=save_html_periodically)
html_thread.daemon = True  # Allow the thread to exit when the main program exits
html_thread.start()

# Find the element containing the property listings
property_listings_div = driver.find_element(By.CSS_SELECTOR, 'div.t-search__list')

# Scroll within the property listings div
scroll_height = 0
while True:
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight;", property_listings_div)
    time.sleep(1)  # Adjust the sleep time as needed
    new_scroll_height = driver.execute_script("return arguments[0].scrollHeight;", property_listings_div)
    if new_scroll_height == scroll_height:
        break
    scroll_height = new_scroll_height

# Find all links that start with '/bolig/'
links = driver.find_elements(By.XPATH, "//a[starts-with(@href, '/bolig/')]")

# Extract and print the links
for link in links:
    print(link.get_attribute('href'))

# Define the on_escape function
def on_escape(event):
    if event.name == 'esc':
        print("Escape key pressed. Exiting...")
        driver.quit()
        exit()

# Register the escape key press event
driver.switch_to.window(driver.current_window_handle)
action_chains = ActionChains(driver)
action_chains.key_down(Keys.ESCAPE).key_up(Keys.ESCAPE).perform()

# Wait for the escape event
driver.switch_to.active_element

# Close the browser (Note: This will not be reached until the escape event occurs)
driver.quit()
```
